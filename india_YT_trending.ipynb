```python
plt.figure(figsize=(10, 6))
sns.histplot(df['view_count'], bins=30, kde=True)
plt.title('Distribution of Video Views')
plt.xlabel('Number of Views')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='likes', y='dislikes')
plt.title('Likes vs. Dislikes')
plt.xlabel('Number of Likes')
plt.ylabel('Number of Dislikes')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])

if numeric_df.empty:
    print("Error: No numeric columns found for correlation calculation.")
else:
    plt.figure(figsize=(10, 6))
    correlation_matrix = numeric_df.corr()
    sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
    plt.title('Correlation Heatmap')
    plt.show()
```

```python
# Q) What is the average number of views per video?
average_views = df['view_count'].mean()
print(f"Average Views per Video: {average_views}")
```

```python
# Q) Which video has the highest number of likes?
most_liked_video = df.loc[df['likes'].idxmax()]
most_liked_video
```

```python
# Q) Is there a trend in the number of views over time?

# Convert 'publish_date' to datetime
df['publish_date'] = pd.to_datetime(df['publish_date'])

# Group by date and sum views
views_over_time = df.groupby(df['publish_date'].dt.date)['view_count'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=views_over_time, x='publish_date', y='view_count')
plt.title('Total Views Over Time')
plt.xlabel('Date')
plt.ylabel('Total Views')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate average views per channel
avg_views_per_channel = df.groupby('channelTitle')['view_count'].mean().reset_index()
avg_views_per_channel = avg_views_per_channel.sort_values(by='view_count', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(data=avg_views_per_channel, x='view_count', y='channelTitle', palette='viridis')
plt.title('Top 10 Channels by Average Views')
plt.xlabel('Average Views')
plt.ylabel('Channel Title')
plt.show()
```

```python
# Calculate likes to dislikes ratio
df['likes_dislikes_ratio'] = df['likes'] / (df['dislikes'] + 1)  # Adding 1 to avoid division by zero

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='view_count', y='likes_dislikes_ratio', alpha=0.6)
plt.axhline(y=df['likes_dislikes_ratio'].mean(), color='r', linestyle='--', label='Average Ratio')
plt.title('Likes to Dislikes Ratio vs. View Count')
plt.xlabel('View Count')
plt.ylabel('Likes to Dislikes Ratio')
plt.legend()
plt.show()
```
